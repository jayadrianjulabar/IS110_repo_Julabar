Rewrite your pay computation with time-and-a-half for overtime and create a function called compute pay which takes two parameters (hours and rate).

    Enter Hours: 45
    Enter Rate: 10
    Pay: 475.0
    
[PEW] E4.6

```python
# Define a function to compute pay
def compute_pay(hours, rate):
    if hours <= 40:
        return hours * rate
    else:
        regular_pay = 40 * rate
        overtime_hours = hours - 40
        overtime_pay = overtime_hours * (1.5 * rate)
        return regular_pay + overtime_pay

# Ask the user to enter hours and rate
hours = float(input("Enter Hours: "))
rate = float(input("Enter Rate: "))

# Call the function and print the result
pay = compute_pay(hours, rate)
print("Pay:", pay)
```

Write a function called computegrade that takes a score as its parameter and returns a grade as a string using the following table:

    Score           Grade
    [ 98, 100]      1.0
    [ 94,  97)      1.25
    [ 90,  93)      1.5
    [ 86,  89)      1.75
    [ 82,  85)      2.0
    [ 78,  81)      2.25
    [ 75,  77)      2.5
    [ 0,   74)      5.0

    Run the program repeatedly to test the various different values for input.
    
Adapted from [PEW] E4.6
<br>

```python
# Function to compute grade based on score
def computegrade(score):
    if 98 <= score <= 100:
        return "1.0"
    elif 94 <= score < 98:
        return "1.25"
    elif 90 <= score < 94:
        return "1.5"
    elif 86 <= score < 90:
        return "1.75"
    elif 82 <= score < 86:
        return "2.0"
    elif 78 <= score < 82:
        return "2.25"
    elif 75 <= score < 78:
        return "2.5"
    elif 0 <= score < 75:
        return "5.0"
    else:
        return "Invalid score"

# Run the program repeatedly for testing
while True:
    try:
        score_input = input("Enter Score (0-100) or 'quit' to exit: ")
        if score_input.lower() == 'quit':
            break
        score = float(score_input)
        grade = computegrade(score)
        print("Grade:", grade)
    except ValueError:
        print("Please enter a valid number.")
```
